### Introduction

In this tutorial we will cover:
- NumPy fundamentals and operations for scientific computing.
- Accelerating NumPy workflows with CuPy on GPUs
- Understanding memory spaces and data movement between CPU and GPU
- Leveraging asynchronous operations for better performance
- Distributed computing with mpi4py for multi-node parallelism

Attendees will be expected to have a general knowledge of HPC and Python programming concepts. Some familiarity with NumPy is helpful but not required.
